```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 1])
plt.show()
```

```python
x=[2,3,5]
y=[1,3,2]
plt.plot(x,y)
plt.title("My First Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.style as style
style.use('ggplot')
x=[2,3,5]
y=[1,3,2]

x2=[1,2,3]
y2=[4,5,1]
plt.plot(x,y, 'g',label='Line 1', linewidth=2)
plt.plot(x2,y2, 'r',label='Line 2', linewidth=2)


plt.title("My First Plot with ggplot style")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True, color='grey')
plt.legend()
plt.show()
```

```python

plt.bar([8, 2, 5,4], [4, 6,8, 1], label="example one", color='blue')

plt.bar([3,1, 6,7], [5, 7, 3,2], label="example two", color='green')

plt.legend()
plt.title('bar graph')
plt.xlabel('number')
plt.ylabel('height')
plt.show()
```

```python
population_ages = [22, 55, 62, 45, 21, 34, 23, 43, 56, 78, 90, 12]
bins = [10, 20, 30, 40, 50, 60, 70, 80, 90]
plt.hist(population_ages, bins, histtype='bar', rwidth=0.8, color='red', label='Age')
plt.xlabel('Ages')
plt.ylabel('Number of People')
plt.title('Population Age Distribution')
plt.legend()
plt.grid(True, color='grey')    
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 1, 4, 7, 2]
plt.scatter(x, y, color='blue', label='Data Points')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True, color='grey')
plt.show()
```

```python
days = [1, 2, 3, 4, 5]
temperatures = [30, 32, 31, 29, 28]
plt.plot(days, temperatures, marker='o', color='orange', label='Temperature')       
plt.title('Temperature Over Days')
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')
plt.xticks(days, ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
plt.grid(True, color='grey')
plt.legend()
plt.show()
```

```python
days = [ 1, 2, 3, 4, 5]
sleeping = [7, 6, 8, 5, 7]
eating = [2, 3, 2, 3, 2]
working = [8, 9, 8, 9, 8]
playing = [7, 6, 6, 7, 7]


plt.plot([], [], color='blue', label='Sleeping', linewidth=5)
plt.plot([], [], color='orange', label='Eating', linewidth=5)
plt.plot([], [], color='green', label='Working', linewidth=5)
plt.plot([], [], color='red', label='Playing', linewidth=5)
plt.title('Daily Activities Over Days')
plt.xlabel('Days')
plt.ylabel('Hours Spent')

plt.stackplot(days, sleeping, eating,playing, working, colors=['blue', 'orange', 'red','green'])
plt.legend(loc='upper left')

plt.grid(True, color='grey')
plt.show()
```

```python
slices = [1, 2, 3, 4, 5]
activities = ['Sleeping', 'Eating', 'Working', 'Playing', 'Other']
cols = ['blue', 'orange', 'green', 'red', 'purple']
plt.pie(slices, labels=activities, autopct='%1.2f%%', startangle=140, colors=cols, explode=(0.1, 0, 0, 0, 0), shadow=True, radius=1.2 )
plt.title('Daily Activities Distribution')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

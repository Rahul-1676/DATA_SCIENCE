```python

print(
    "Highest CO2 Emission per capita(tons) in",
    df.loc[df['CO2 Emission per Capita (tons)'].idxmax(), 'Year'],
    "that is",
    df['CO2 Emission per Capita (tons)'].max()
)
print(
    "Lowest CO2 Emission per capita(tons) in",
    df.loc[df['CO2 Emission per Capita (tons)'].idxmin(), 'Year'],
    "that is",
    df['CO2 Emission per Capita (tons)'].min()
)
```

```python
print("Heavy rainfall(mm) in", 
      df.loc[df['Rainfall (mm/year)'].idxmax(), 'Year'],
      "that is",
      df['Rainfall (mm/year)'].max()
      )
print("low rainfall (mm) in",
      df.loc[df['Rainfall (mm/year)'].idxmin(), 'Year'],
      "that is",
      df['Rainfall (mm/year)'].min()
      )
```
